```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_coordinates(df, start_date_str, end_date_str):
    # Assuming the date format in the CSV is '1402-12-09'
    # If the format is different, adjust accordingly
    start_date = start_date_str
    end_date = end_date_str

    names = df['name'].unique()
    dfs = []
    for name in names:
        df_temp = df[df['name'] == name]
        df_temp = df_temp[(df_temp['date'] >= start_date) & (df_temp['date'] <= end_date)]
        dfs.append(df_temp)
    plt.figure(figsize=(6, 4))
    for i, df_temp in enumerate(dfs):
        plt.plot(df_temp['date'], df_temp['sale'], label=names[i])

    plt.xlabel('date')
    plt.ylabel('sale')
    plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.xticks(rotation=90)
    plt.show()

# Load your dataset (assuming 'dataset.csv' is in the same directory)
df = pd.read_csv('dataset.csv')


```

```python
# Define start_date and end_date as strings
start_date = '1402-12-08'
end_date = '1402-12-15'

# Call the function with the corrected date arguments
plot_coordinates(df, start_date, end_date)
```

```python

def draw_graph(df_merged, start_date_str, end_date_str):
    # Read the csv file into a pandas dataframe
    
    start_date = start_date_str
    end_date = end_date_str
    # Filter the dataframe based on the start and end dates
    filtered_df = df_merged[(df_merged['date'] >= start_date) & (df_merged['date'] <= end_date)]
    
    # Plot the graph
    plt.plot(filtered_df['date'], filtered_df['sale'])
    plt.xlabel('Date')
    plt.ylabel('Sale')
    plt.title('Sale over Time')
    plt.grid(color='green', linestyle='--', linewidth=0.5)
    plt.xticks(rotation=90)
    plt.show()
```
